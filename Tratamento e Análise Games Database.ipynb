## 🎮 Tratamento de qualidade do Dataset

```python
df.info()
```

```python
# Das colunas existentes no banco de dados, inicialmente excluo-as pois não será relevante para a análise inicial
# São elas: IMG(Arquivo de imagem), Critic Score e Last Update.

df = df.drop(columns=['img', 'last_update', 'critic_score'])
```

```python
df.info()
```

```python
# O registro de numero 11 consta como data e a tipagem está como string. Converter usando:

pd.to_datetime(df['release_date'], format='%d-%m-%Y')
```

```python
df.info()
```

```python
df.dtypes
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], format= '%d-%m-%Y')
```

```python
df.dtypes
```

```python
# Depois de excluido inciialmente três colunas, vamos tratar os nulos caso a caso:
# a coluna developer, como é texto e não influenciará diretamente nas estatísticas, preencho com Desconhecido usando o comando abaixo:

df['developer'] = df['developer'].fillna('Desconhecido')
```

```python
df['developer'].isnull().sum()
```

```python
# Criando uma coluna específica com a informação do ano apenas para facilitar a análise temporal

df['year'] = df['release_date'].dt.year
```

```python
# De todo dataset preciso realizar filtro dos jogos com informações de vendas que terá uma subanálise só sobre
# Essas informações. a df geral utiizarei para outras investigações que não dependem de dados de vendas.

df_vendas = df.dropna(subset=['total_sales'])
```

```python
print("Dataset original:", df.shape)
print("Com vendas:", df_vendas.shape)
print("Nulos restantes:")
print(df_vendas.isnull().sum())
```
