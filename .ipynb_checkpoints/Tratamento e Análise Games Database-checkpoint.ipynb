## 🎮 Tratamento de qualidade do Dataset

```python
df.info()
```

```python
# Das colunas existentes no banco de dados, inicialmente excluo-as pois não será relevante para a análise inicial
# São elas: IMG(Arquivo de imagem), Critic Score e Last Update.

df = df.drop(columns=['img', 'last_update', 'critic_score'])
```

```python
df.info()
```

```python
# O registro de numero 11 consta como data e a tipagem está como string. Converter usando:

pd.to_datetime(df['release_date'], format='%d-%m-%Y')
```

```python
df.info()
```

```python
df.dtypes
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], format= '%d-%m-%Y')
```

```python
df.dtypes
```

```python
# Depois de excluido inciialmente três colunas, vamos tratar os nulos caso a caso:
# a coluna developer, como é texto e não influenciará diretamente nas estatísticas, preencho com Desconhecido usando o comando abaixo:

df['developer'] = df['developer'].fillna('Desconhecido')
```

```python
df['developer'].isnull().sum()
```

```python
# Criando uma coluna específica com a informação do ano apenas para facilitar a análise temporal

df['year'] = df['release_date'].dt.year
```

```python
# De todo dataset preciso realizar filtro dos jogos com informações de vendas que terá uma subanálise só sobre
# Essas informações. a df geral utiizarei para outras investigações que não dependem de dados de vendas.

df_vendas = df.dropna(subset=['total_sales']).copy()
```

```python
print("Dataset original:", df.shape)
print("Com vendas:", df_vendas.shape)
print("Nulos restantes:")
print(df_vendas.isnull().sum())
```

```python
df_vendas[['na_sales', 'jp_sales', 'pal_sales', 'other_sales']].isnull().sum()
```

## 🎮 Análise das 10 perguntas à investigar com as filtragens, agrupamentos, gráficos e etc.

### 1. Quais o gêneros mais produzidos?

```python
generos = df['genre'].value_counts()
print(generos)
```

```python
df['genre'].value_counts().head(10).plot(kind='bar')
```

### 2. Qual Desenvolvedor domina cada gênero?

```python
contagem = df.groupby(['genre', 'developer']).size().reset_index(name='total_jogos')
print(contagem)
```

```python
mais_generos = contagem.loc[contagem.groupby('genre')['total_jogos'].idxmax()]
print(mais_generos)
```

```python
# Juntar gênero + developer no gráfico
mais_generos['label'] = mais_generos['genre'] + ' — ' + mais_generos['developer']

mais_generos.plot(
    kind='barh',
    x='label',
    y='total_jogos',
    title='Publisher dominante por gênero',
    figsize=(10, 8),
    legend=False
)
```

### 3. Qual gênero é mais popular no Japão?

```python
# Filtrar os jogos com registro de vendas no Japão


df_jp = df_vendas.dropna(subset=['jp_sales'])
```

```python
# Agrupo as vendas por gênero

vendas_jp = df_jp.groupby('genre')['jp_sales'].sum().sort_values(ascending=False)
print(vendas_jp)
```

```python
# Plotando o gráfico com o resultado

vendas_jp.head(10).plot(
    kind='barh',
    title='Gêneros mais vendidos no Japão',
    figsize=(10, 6)
)
```

### 4. Qual console tem o maior catálogo de jogos?

```python
catalogo = df['console'].value_counts().head(10)
```

```python
print(catalogo)
```

```python
catalogo.plot(
    kind='bar',
    title='Top 10 consoles com maior catálogo de jogos',
    figsize=(10, 6)
)
```

### 5.Qual o top 10 de jogos mais vendidos de todos os tempos?

```python
#Aqui trabalhar com a base df_vendas por se tratar de análise de vendas

top10_total = df_vendas.groupby('title')['total_sales'].sum().nlargest(10)
```

```python
top10_total.plot(
    kind='bar',
    title='Top 10 jogos — vendas somadas (todas plataformas)',
    figsize=(10, 6)
)
```

### 6.Qual console tem maior media de vendas por jogo?

```python
media_console = df_vendas.groupby('console')['total_sales'].mean().sort_values(ascending=False)
```

```python
print(media_console)
```

```python
#Para evitar outlier e que distorça a média, aplico uma regra de número mínimo de catálogos por console. 
#Começo contando os catálogos por console

contagem = df_vendas.groupby('console')['total_sales'].count()
```

```python
#Aplico a regra mínima de catálogo:

consoles_considerados = contagem[contagem>=50].index
```

```python
# Agora reaplico a média

nova_media= df_vendas[df_vendas['console'].isin(consoles_considerados)].groupby('console')['total_sales'].mean().sort_values(ascending=False)
```

```python
#Resultado plotado no gráfico

nova_media.head(15).plot(
    kind='barh',
    title='Média de vendas por jogo (consoles com 50+ títulos)',
    figsize=(10, 6)
)
```

### 7.Qual genero mais popular na America do Norte?


```python
# Filtrar os jogos com registro de vendas na America do Norte


df_na = df_vendas.dropna(subset=['na_sales'])
```

```python
# Agrupo as vendas por gênero

vendas_na = df_na.groupby('genre')['na_sales'].sum().sort_values(ascending=False)
print(vendas_na)
```

```python
# Plotando o gráfico com o resultado

vendas_na.head(10).plot(
    kind='bar',
    title='Gêneros mais vendidos na America do Norte',
    figsize=(10, 6)
)
```

### 8. Como as vendas evoluíram por decada?


```python
#Criar uma coluna com década, parecido com a coluna Year que criei no início do Nb

df_vendas['decada'] = (df_vendas['year'] // 10)*10
```

```python
#Total de vendas por década

vendas_decada = df_vendas.groupby('decada')['total_sales'].sum()
```

```python
print(vendas_decada)
```

```python
#Como o dataset foca em mídia física.. nota-se a queda no gráfico abaixo devido ao boom das mídias digitais, 
#por exemplo
#MÍDIAS FÍSICAS ESTÃO MORRENDO.

vendas_ano = df_vendas.groupby('year')['total_sales'].sum()

vendas_ano.plot(
    kind='line',
    title='Evolução de vendas por ano',
    figsize=(12, 6)
)
```

### 9. Quais os títulos mais vendidos no seculo XX?

```python
#Com a coluna de Year e Decadas criada, agora realizo o filtro da informação que desejo
# Século passado: 1980 a 2000
#Depois do filtro aplico meu top10


seculo_xx = df_vendas[(df_vendas['year'] >= 1980) & (df_vendas['year'] <= 2000)]
```

```python
top10_xx = seculo_xx.groupby('title')['total_sales'].sum().nlargest(10)
```

```python
print(top10_xx)
```

```python
top10_xx.plot(
    kind='bar',
    title='Top 10 jogos mais vendidos (1980-2000)',
    figsize=(10, 6)
)
```

### 10. Quais os títulos mais vendidos no seculo XXI?

```python
# Século XXI: 2001 em diante

seculo_xxi = df_vendas[df_vendas['year'] >= 2001]
```

```python
top10_xxi = seculo_xxi.groupby('title')['total_sales'].sum().nlargest(10)
```

```python
print(top10_xxi)
```

```python
#Plot da informação

top10_xxi.plot(
    kind='bar',
    title='Top 10 jogos mais vendidos (2001+)',
    figsize=(10, 6)
)
```
